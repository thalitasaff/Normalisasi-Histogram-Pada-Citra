**CITRA MUSHROOM 1**

```python
img1 = cv2.imread('C:\\Users\ASUS\Analisis Data Image\mushroom1.jpeg',1)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

# compute a grayscale histogram
hist1 = cv2.calcHist([img1], [0], None, [256], [0, 256])
```

```python
# matplotlib expects RGB images so convert and then display the image
# with matplotlib
plt.figure()
plt.axis("off")
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_GRAY2RGB))
# plot the histogram
plt.figure()
plt.title("Mushroom 1 Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(hist1)
plt.xlim([0, 256])
```

```python
# NORMALISASI HISTORAM CITRA MUSHROM 1
# compute a grayscale histogram
hist1 /= hist1.sum()
# plot the normalized histogram
plt.figure()
plt.title("Mushroom 1 Histogram (Normalized)")
plt.xlabel("Bins")
plt.ylabel("% of Pixels")
plt.plot(hist1)
plt.xlim([0, 256])
plt.show()
```
